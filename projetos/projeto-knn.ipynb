```python
# Função para calcular a distância

def calcula_distancia(ponto1, ponto2):
    try:
        soma = 0
        quadrados = tuple((elem1 - elem2)**2 for elem1, elem2 in zip(ponto1, ponto2)) # Calcula o quadrado
        for elemento in quadrados:
            soma += elemento  #soma de quadrados
            
        return soma ** 0.5  
    except Exception as e:
        print("Erro ao calcular a distância:", e)
        return None
```

```python
# Função para classificar os dados
def classifica_dados(data, no_class, k):
    previsoes = []
    
    # Laço para cada item de no_class faz o cálculo da distancia com todos os itens de data
    for cliente_desconhecido in no_class:
        distancias = []
        for cliente_conhecido in data:
            dist = calcula_distancia(cliente_desconhecido[2], cliente_conhecido[2])
            distancias.append((cliente_conhecido, dist))
            
        # ordenação pela distancia    
        distancias.sort(key=lambda x: x[1]) 
        
        
        # Lista com os vizinhos
        vizinhos = []
        #Percorre K elementos da lista distancias de acordo com o parametro K
        for i in range(k):
            k_vizinhos = distancias[i]
            vizinhos.append(k_vizinhos[0])
              
        # Classificação dos vizinhos,a lista ocorrencias vai possuir a classificação
        # e o respectivo número de ocorrências
        ocorrencias = []

        for vizinho in vizinhos:
            classificacao = vizinho[1]
            classe_existente = False
        # Percorre a lista ocorrencias incrementa se a classificacao já existir na lista, caso contrário adiciona
            for i, c in enumerate(ocorrencias):
                if c[0] == classificacao:          
                    ocorrencias[i][1] += 1
                    classe_existente = True
                    break
            if not classe_existente:
                ocorrencias.append([classificacao, 1])

        # Obtem a classificação conforme o máximo de ocorrencias            
        classificacao_prevista = max(ocorrencias, key=lambda x: x[1])[0]
        
        previsoes.append((cliente_desconhecido[0], classificacao_prevista,cliente_desconhecido[2]))
    
 
    return previsoes

```

```python
# Definindo o valor de k
k = 5
# Retorna a lista com as classificações
dados_classificados = classifica_dados(data, no_class, k)
for classificacao in dados_classificados:
    print(classificacao)
  
```
